```python
import numpy as np
from scipy import signal
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import random
import scipy.stats as stats
import matplotlib.mlab as mlab 
import matplotlib.gridspec as gridspec
import math

gs = gridspec.GridSpec(4, 6)
plt.rcParams['figure.figsize'] = (10.5, 6)

plt.subplot(gs[:2,:2])
mu = 0
sigma = np.sqrt(3*(np.pi)**2/128)
u = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

score = []
with open("Template A_t3_abs_100_1000.txt", "r") as f:
    for line in f:
        score.append(float(line.strip()))
score = [i*(100**0.5) for i in score]        
plt.hist(score, bins=38, normed =1)
plt.plot(u, stats.norm.pdf(u, mu, sigma))  
plt.ylabel('Probability density') 
plt.xlabel(r'$\sqrt{n}(\hat{\theta_n}-\theta^*$)',fontsize =11) 
plt.title('n=100') 


plt.subplot(gs[:2,2:4])
mu = 0
sigma = np.sqrt(3*(np.pi)**2/128)
u = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

score = []
with open("Template A_t3_abs_500_1000.txt", "r") as f:
    for line in f:
        score.append(float(line.strip()))
score = [i*(500**0.5) for i in score]        
plt.hist(score, bins=38, normed =1)
plt.plot(u, stats.norm.pdf(u, mu, sigma))  
plt.ylabel('Probability density') 
plt.xlabel(r'$\sqrt{n}(\hat{\theta_n}-\theta^*$)',fontsize =11) 
plt.title('n=500') 


plt.subplot(gs[:2,4:6])
mu = 0
sigma = np.sqrt(3*(np.pi)**2/128)
u = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

score = []
with open("Template A_t3_abs_1000_1000.txt", "r") as f:
    for line in f:
        score.append(float(line.strip()))
score = [i*(1000**0.5) for i in score]        
plt.hist(score,bins=38, normed =1)
plt.plot(u, stats.norm.pdf(u, mu, sigma))  
plt.ylabel('Probability density') 
plt.xlabel(r'$\sqrt{n}(\hat{\theta_n}-\theta^*$)',fontsize =11) 
plt.title('n=1000')  

plt.subplot(gs[2:, 1:3])
mu = 0
sigma = np.sqrt(3*(np.pi)**2/128)
u = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

score = []
with open("Template A_t3_abs_5000_1000.txt", "r") as f:
    for line in f:
        score.append(float(line.strip()))
score = [i*(5000**0.5) for i in score]        
plt.hist(score, bins=38, normed =1)
plt.plot(u, stats.norm.pdf(u, mu, sigma))  
plt.ylabel('Probability density')
plt.xlabel(r'$\sqrt{n}(\hat{\theta_n}-\theta^*$)',fontsize =11) 
plt.title('n=5000') 


plt.subplot(gs[2:, 3:5])
mu = 0
sigma = np.sqrt(3*(np.pi)**2/128)
u = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

score = []
with open("Template A_t3_abs_10000_1000.txt", "r") as f:
    for line in f:
        score.append(float(line.strip()))
score = [i*100 for i in score]        
plt.hist(score, bins=38, normed =1)
plt.plot(u, stats.norm.pdf(u, mu, sigma))  
plt.ylabel('Probability density') 
plt.xlabel(r'$\sqrt{n}(\hat{\theta_n}-\theta^*$)',fontsize =11) 
plt.title('n=10000') 

plt.tight_layout()
plt.show() 
```
