```python
import numpy as np
from scipy import signal
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import random
start_time = datetime.datetime.now() 

n = 1000
x = np.random.rand(n)
x = np.sort(x)
e = np.random.normal(0, 1, n)
y = [e[i] if (x[i] <=0.25 or x[i]>=0.5) else 4*x[i]-1+e[i] for i in range(n)]
f = [0 if (x[i] <= (0.25) or x[i]>=(0.5)) else 4*(x[i])-1 for i in range(n)]
```

```python
plt.rcParams['figure.figsize'] = (6.0, 8.0)

plt.subplot(2,2,1)
plt.plot(x,f, c = 'black', label = 'Filer')
plt.title('Filter')

plt.subplot(2,2,2)
plt.scatter(x,y, c='gray', alpha=0.5, s=5, label = 'Signal')
plt.plot(x,f, c = 'black', label = 'Filer')
plt.ylim(-5, 5)
plt.title('Normal noise')  

plt.subplot(2,2,3)
e = np.random.standard_t(3, size=n)
y = [e[i] if (x[i] <=0.2 or x[i]>=0.8 or (x[i]>=0.4 and x[i]<=0.6)) else 1+e[i] for i in range(n)]
plt.scatter(x,y, c='gray', alpha=0.5, s=5, label = 'Signal')
plt.plot(x,f, c = 'black', label = 'Filer')
plt.ylim(-5, 5)
plt.title('$T_3$ noise') 

plt.subplot(2,2,4)
e = np.random.standard_cauchy(n)
y = [e[i] if (x[i] <=0.2 or x[i]>=0.8 or (x[i]>=0.4 and x[i]<=0.6)) else 1+e[i] for i in range(n)]
plt.scatter(x,y, c='gray', alpha=0.5, s=5, label = 'Signal')
plt.plot(x,f, c = 'black', label = 'Filer')
plt.legend(bbox_to_anchor=(1.02, 0), loc=3, borderaxespad=0)
plt.ylim(-5, 5)
plt.title('Cauchy noise') 

plt.tight_layout()
plt.show() 
```
