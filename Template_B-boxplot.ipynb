```python
import numpy as np
from scipy import signal
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import random

plt.rcParams['figure.figsize'] = (13.5, 7)

plt.subplot(1,3,1)
score1 = []
with open("Template B_normal1_square_10000_200.txt", "r") as f:
    for line in f:
        score1.append(float(line.strip()))
score2 = []
with open("Template B_normal1_abs_10000_200.txt", "r") as f:
    for line in f:
        score2.append(float(line.strip()))
score3 = []
with open("Template B_normal1_huber_10000_200.txt", "r") as f:
    for line in f:
        score3.append(float(line.strip()))
score4 = []
with open("Template B_normal1_tukey_10000_200.txt", "r") as f:
    for line in f:
        score4.append(float(line.strip()))
        
score1 = [abs(i) for i in score1]
score2 = [abs(i) for i in score2]
score3 = [abs(i) for i in score3]
score4 = [abs(i) for i in score4]
all_data = [score1, score2, score3, score4]
labels = ["Squared error", "Absolute-value", "HUber","Tukey"]
bplot = plt.boxplot(all_data, patch_artist=True, labels=labels)  
plt.yscale('log')
plt.title('Normal noise',fontsize=14)
colors = ['pink', 'lightblue', 'blue', 'green']
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)  
plt.xticks(rotation=30,fontsize=14)     
plt.yticks(fontsize=14)
plt.ylabel(r'Estimation error $|\hat{\theta_n}-\theta^*|$',fontsize=14) 


plt.subplot(1,3,2)
score1 = []
with open("Template B_t3_square_10000_200.txt", "r") as f:
    for line in f:
        score1.append(float(line.strip()))
score2 = []
with open("Template B_t3_abs_10000_200.txt", "r") as f:
    for line in f:
        score2.append(float(line.strip()))
score3 = []
with open("Template B_t3_huber_10000_200.txt", "r") as f:
    for line in f:
        score3.append(float(line.strip()))
score4 = []
with open("Template B_t3_tukey_10000_200.txt", "r") as f:
    for line in f:
        score4.append(float(line.strip()))

score1 = [abs(i) for i in score1]
score2 = [abs(i) for i in score2]
score3 = [abs(i) for i in score3]
score4 = [abs(i) for i in score4]
all_data = [score1, score2, score3, score4]
labels = ["Squared error", "Absolute-value", "HUber","Tukey"]
bplot = plt.boxplot(all_data, patch_artist=True, labels=labels) 
plt.yscale('log')
plt.title('T3 noise',fontsize=14)
colors = ['pink', 'lightblue', 'blue', 'green']
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)  
plt.xticks(rotation=30,fontsize=14)     
plt.yticks(fontsize=14)

plt.subplot(1,3,3)
score1 = []
with open("Template B_cauchy1_square_10000_200.txt", "r") as f:
    for line in f:
        score1.append(float(line.strip()))
score2 = []
with open("Template B_cauchy1_abs_10000_200.txt", "r") as f:
    for line in f:
        score2.append(float(line.strip()))
score3 = []
with open("Template B_cauchy1_huber_10000_200.txt", "r") as f:
    for line in f:
        score3.append(float(line.strip()))
score4 = []
with open("Template B_cauchy1_tukey_10000_200.txt", "r") as f:
    for line in f:
        score4.append(float(line.strip()))

score1 = [abs(i) for i in score1]
score2 = [abs(i) for i in score2]
score3 = [abs(i) for i in score3]
score4 = [abs(i) for i in score4]
all_data = [score1, score2, score3, score4]
labels = ["Squared error", "Absolute-value", "HUber","Tukey"]
bplot = plt.boxplot(all_data, patch_artist=True, labels=labels) 
plt.yscale('log')
plt.title('Cauchy noise',fontsize=14)
colors = ['pink', 'lightblue', 'blue', 'green']
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)  
plt.xticks(rotation=30,fontsize=14)     
plt.yticks(fontsize=14)
plt.show()
```
