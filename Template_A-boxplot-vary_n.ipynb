```python
import numpy as np
from scipy import signal
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import random

plt.rcParams['figure.figsize'] = (13.5, 7)

score0 = []
with open("Template A_t3_abs_100_1000.txt", "r") as f:
    for line in f:
        score0.append(float(line.strip()))
score1 = []
with open("Template A_t3_abs_500_1000.txt", "r") as f:
    for line in f:
        score1.append(float(line.strip()))
score2 = []
with open("Template A_t3_abs_1000_1000.txt", "r") as f:
    for line in f:
        score2.append(float(line.strip()))
score3 = []
with open("Template A_t3_abs_5000_1000.txt", "r") as f:
    for line in f:
        score3.append(float(line.strip()))
score4 = []
with open("Template A_t3_abs_10000_1000.txt", "r") as f:
    for line in f:
        score4.append(float(line.strip()))

score0 = [abs(i) for i in score0]
score1 = [abs(i) for i in score1]
score2 = [abs(i) for i in score2]
score3 = [abs(i) for i in score3]
score4 = [abs(i) for i in score4]
all_data = [score0, score1, score2, score3, score4]
labels = ["100","500", "1000", "5000","10000"]
bplot = plt.boxplot(all_data, patch_artist=True, labels=labels) 
plt.yscale('log')

colors = ['orange','pink', 'lightblue', 'blue', 'green']
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)  
plt.xticks(fontsize=20)     
plt.yticks(fontsize=20)
plt.ylabel(r'Estimation error $|\hat{\theta_n}-\theta^*|$',fontsize=20) 
plt.xlabel('n',fontsize=20)
```
