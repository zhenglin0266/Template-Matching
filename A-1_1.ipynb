```python
import numpy as np
from scipy import signal
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import random
start_time = datetime.datetime.now() 

n = 10000
x = np.random.rand(n)
x = np.sort(x)
e = np.random.normal(0, 1, n)
y1 = [(1-x[i]**2)**3 for i in range(n)] 
y = [e[i]+y1[i] if y1[i]>0 else e[i] for i in range(n)]
f1 = [(1-(x[i])**2)**3 for i in range(n)]  
f = [f1[i] if f1[i]>0 else 0 for i in range(n)]
```

```python
plt.rcParams['figure.figsize'] = (6.0, 5.0)

plt.subplot(4,1,1)
plt.plot(x,f, c = 'black', label = 'Filer')
plt.title('Filter') 

plt.subplot(4,1,2)
plt.scatter(x,y, c='gray', alpha=0.5, s=10, label = 'Signal')
plt.plot(x,f, c = 'black', label = 'Filer')
plt.title('Normal noise') 

plt.subplot(4,1,3)
e = np.random.standard_t(3, size=n)
y1 = [(1-x[i]**2)**3 for i in range(n)] 
y = [e[i]+y1[i] if y1[i]>0 else e[i] for i in range(n)]
plt.scatter(x,y, c='gray', alpha=0.5, s=10, label = 'Signal')
plt.plot(x,f, c = 'black', label = 'Filer')
plt.title('T3 noise') 

plt.subplot(4,1,4)
e = np.random.standard_cauchy(n)
y1 = [(1-x[i]**2)**3 for i in range(n)] 
y = [e[i]+y1[i] if y1[i]>0 else e[i] for i in range(n)]
plt.scatter(x,y, c='gray', alpha=0.5, s=10, label = 'Signal')
plt.plot(x,f, c = 'black', label = 'Filer')
plt.legend(bbox_to_anchor=(1.02, 0), loc=3, borderaxespad=0)
plt.title('Cauchy noise') 

plt.tight_layout()
plt.show() 
```
