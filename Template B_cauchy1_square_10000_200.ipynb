```python
import numpy as np
from scipy import signal
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import random
import math
start_time = datetime.datetime.now()

n = 10000
m = int(n/2)
err_res = []
for j in range(200):
    x = np.random.rand(n)
    x = np.sort(x)
    e = np.random.standard_cauchy(n)
    y1 = [(1-(4*(x[i]-0.5))**2)**3 for i in range(n)] 
    y = [e[i]+y1[i] if y1[i]>0 else e[i] for i in range(n)]
    res = []
    for a in range(-m,m):
        f1 = [(1-(4*(x[i]-0.5-a/n))**2)**3 for i in range(n)] 
        f = [f1[i] if f1[i]>0 else 0 for i in range(n)]         
        diff = sum([(y[i]-f[i])**2 for i in range(n)])
        res.append(diff)
    min_res = min(res)
    in_res = [i for i in range(2*m) if res[i]==min_res]
    lo_res = random.sample(in_res, 1)[0]
    err = (lo_res-m)/m
    err_res.append(err)
    print(j)

with open("Template B_cauchy1_square_10000_200.txt", "w") as f:
    for s in err_res:
        f.write(str(s) +"\n")  

        
# score = []
# with open("file.txt", "r") as f:
#     for line in f:
#         score.append(float(line.strip()))
   
end_time = datetime.datetime.now()    
interval = (end_time-start_time).seconds
final_time = interval/60.0
print ('Time:\t', final_time)   
```
