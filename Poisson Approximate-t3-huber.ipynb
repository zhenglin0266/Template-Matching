```python
import numpy as np
from scipy import signal
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import random
import math
start_time = datetime.datetime.now() 

def huber(s,c):
    if abs(s)>c:
        return c*abs(s)-c**2/2
    else:
        return s**2/2

n = 10000
m= int(n/50)
err_res = []
for j in range(200):
    x = np.random.rand(n)
    x = np.sort(x)
    e = np.random.standard_t(3,size=n)
    res = []
    for i in range(-m,0):
        res.append(np.sum([huber(e[k]+1,0.5)-huber(e[k],0.5) if x[k] <= 0.5 and x[k]>= 0.5+i/n else 0 for k in range(n)]))

    for i in range(m):
        res.append(np.sum([huber(e[k]+1,0.5)-huber(e[k],0.5) if x[k] >= 0.5 and x[k] < 0.5+i/n else 0 for k in range(n)]))
        
    min_res = min(res)
    in_res = [i for i in range(2*m) if res[i]==min_res]
    lo_res = random.sample(in_res, 1)[0]
    err = lo_res-m
    err_res.append(err)
    print(j)
    
with open("Poisson Approximate_E-t3-huber.txt", "w") as f:
    for s in err_res:
        f.write(str(s) +"\n")   
        
    
end_time = datetime.datetime.now()     
interval = (end_time-start_time).seconds
final_time = interval/60.0 
print ('Time:\t', final_time)   
```
