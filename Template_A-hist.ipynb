```python
import numpy as np
from scipy import signal
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import random
import scipy.stats as stats
import matplotlib.mlab as mlab 
import math

plt.rcParams['figure.figsize'] = (10.5, 3)

plt.subplot(1,4,1)
mu = 0
sigma = np.sqrt(1/8)
u = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

score = []
with open("Template A_normal1_square_10000_200.txt", "r") as f:
    for line in f:
        score.append(float(line.strip()))
score = [i*100 for i in score]        
plt.hist(score, normed =1)
plt.plot(u, stats.norm.pdf(u, mu, sigma))  
plt.ylabel('Probability density') 
plt.xlabel(r'$\sqrt{n}(\hat{\theta_n}-\theta^*$)',fontsize =11) 
plt.title('Squared error') 

plt.subplot(1,4,2)
mu = 0
sigma = np.sqrt(np.pi/16)
u = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

score = []
with open("Template A_normal1_abs_10000_200.txt", "r") as f:
    for line in f:
        score.append(float(line.strip()))
score = [i*100 for i in score]        
plt.hist(score, normed =1)
plt.plot(u, stats.norm.pdf(u, mu, sigma))  
plt.ylabel('Probability density') 
plt.xlabel(r'$\sqrt{n}(\hat{\theta_n}-\theta^*$)',fontsize =11) 
plt.title('Absolute-value')  

plt.subplot(1,4,3)
mu = 0
sigma = np.sqrt(1.03709/8)
u = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

score = []
with open("Template A_normal1_huber_10000_200.txt", "r") as f:
    for line in f:
        score.append(float(line.strip()))
score = [i*100 for i in score]        
plt.hist(score, normed =1)
plt.plot(u, stats.norm.pdf(u, mu, sigma))  
plt.ylabel('Probability density') 
plt.xlabel(r'$\sqrt{n}(\hat{\theta_n}-\theta^*$)',fontsize =11) 
plt.title('Huber') 

plt.subplot(1,4,4)
mu = 0
sigma = np.sqrt(1.2941/8)
u = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

score = []
with open("Template A_normal1_tukey_10000_200.txt", "r") as f:
    for line in f:
        score.append(float(line.strip()))
score = [i*100 for i in score]        
plt.hist(score, normed =1)
plt.plot(u, stats.norm.pdf(u, mu, sigma))  
plt.ylabel('Probability density') 
plt.xlabel(r'$\sqrt{n}(\hat{\theta_n}-\theta^*$)',fontsize =11) 
plt.title('Tukey') 

plt.tight_layout()
plt.show() 
```
