```python
import numpy as np
from scipy import signal
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import random
start_time = datetime.datetime.now() 

n = 5000
m = int(n/10)
err_res = []
for j in range(1000):
    x = np.random.rand(n)
    x = np.sort(x)
    e = np.random.standard_t(3, size=n)
    y = [e[i] if (x[i] <=0.25 or x[i]>=0.5) else 4*x[i]-1+e[i] for i in range(n)]
    res = []
    for a in range(-m,m):
        f = [0 if (x[i] <= (0.25+a/(10*n)) or x[i]>=(0.5+a/(10*n))) else 4*(x[i]-a/(10*n))-1 for i in range(n)]
        diff = sum([abs(y[i]-f[i]) for i in range(n)])
        res.append(diff)
    min_res = min(res)
    in_res = [i for i in range(2*m) if res[i]==min_res]
    lo_res = random.sample(in_res, 1)[0]
    err = (lo_res-m)/(10*n)
    err_res.append(err)
    print(j)

with open("Template E_t3_abs_5000_1000.txt", "w") as f:
    for s in err_res:
        f.write(str(s) +"\n")   
        
# score = []
# with open("file.txt", "r") as f:
#     for line in f:
#         score.append(float(line.strip()))
    
end_time = datetime.datetime.now()     
interval = (end_time-start_time).seconds
final_time = interval/60.0 
print ('Time:\t', final_time)   
```
