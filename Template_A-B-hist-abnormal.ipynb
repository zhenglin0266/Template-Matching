```python
import numpy as np
from scipy import signal
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import random
import scipy.stats as stats
import matplotlib.mlab as mlab 
import math


plt.rcParams['figure.figsize'] = (10.5, 3)

plt.subplot(1,2,1)

score = []
with open("Template A_cauchy1_square_10000_200.txt", "r") as f:
    for line in f:
        score.append(float(line.strip()))
score = [i*100 for i in score]        
plt.hist(score, normed =1)
plt.ylabel('Probability density') 
plt.xlabel(r'$\sqrt{n}(\hat{\theta_n}-\theta^*$)',fontsize =11) 
plt.title('Filter A in Cauchy noise (Squared error)')  

plt.subplot(1,2,2)

score = []
with open("Template B_cauchy1_square_10000_200.txt", "r") as f:
    for line in f:
        score.append(float(line.strip()))
score = [i*100 for i in score]        
plt.hist(score, normed =1) 
plt.ylabel('Probability density')
plt.xlabel(r'$\sqrt{n}(\hat{\theta_n}-\theta^*$)',fontsize =11) 
plt.title('Filter B in Cauchy noise (Squared error)') 

plt.tight_layout()
plt.show() 
```
